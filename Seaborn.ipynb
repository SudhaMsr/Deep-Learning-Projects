```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
plt.style.use('seaborn-bright')
```

```python
sns.get_dataset_names()
```

## Plots
- All the basic plots
- Heatmap
- Regression plot
- joint plots
- count plot

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe()
```

### Distribution plots

```python
sns.distplot(tips.tip)
plt.show()
```

```python
sns.distplot(tips.tip,kde=False)
plt.show()
```

```python
sns.kdeplot(tips.tip)
plt.show()
```

### Box plot

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.show()
```

### Line plot

```python
sns.lineplot(x='size',y='tip',data=tips)
plt.show()
```

### Scatter plot

```python
tips.head(2)
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='day',style='time',size='size')
plt.show()
```

### Bar plot

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.show()
```

### Regression plot

```python
sns.regplot(x='total_bill',y='tip',data=tips)
plt.show()
```

### Violin plot

```python
plt.figure(figsize=(10,10))
sns.catplot(x='day',y='tip',data=tips,kind='violin')
plt.show()
```

### Joint plot

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
plt.show()
```

## Count plot

```python
tips.head()
```

```python
tips.sex.value_counts()
```

```python
tips.smoker.value_counts()
```

```python
tips.day.value_counts()
```

```python
tips.time.value_counts()
```

```python
sns.countplot(x='day',data=tips)
plt.show()
```

```python
sns.countplot(x='day',data=tips,hue='time')
plt.show()
```

### Pair plot

```python
sns.pairplot(tips,hue='time') # useful if few numerical variables are to be analysed
plt.show()
```

```python
tips.corr()
```

```python
# Heatmap is not constrained by no of numerical variables like the pairplot
plt.figure(figsize=(10,5)) # very useful plot to check correlations between any two numerical variables
sns.heatmap(tips.corr(),annot=True)
plt.show()
```

### Relational plot

```python
sns.relplot(x='total_bill',y='tip',data=tips,col='day',row='smoker')
plt.show()
```

### Subplots

```python
plt.figure(figsize=(10,10))
plt.subplot(2,1,1) # row,column,position of plot
sns.countplot(x='day',data=tips)

plt.subplot(2,1,2)
sns.heatmap(tips.corr(),annot=True)
plt.show()
```
