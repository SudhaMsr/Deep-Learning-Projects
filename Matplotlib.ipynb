#### Setting seed for reproducing the same plot

```python
np.random.seed(1)
temperature = np.random.uniform(18,43,size=30)
plt.plot(temperature)
plt.show()
```

#### Modifying the size of the chart

```python
plt.figure(figsize=(10,5))
np.random.seed(1)
temperature = np.random.uniform(18,43,size=30)
plt.plot(temperature)
plt.show()
```

```python
# Plotting multiple line charts
plt.figure(figsize=(15,5))
np.random.seed(1)
summer = np.random.uniform(25,45,size=30)
winter = np.random.uniform(10,30,size=30)
rainy = np.random.uniform(20,40,size=30)
plt.plot(summer)
plt.plot(winter)
plt.plot(rainy)
plt.show()
```

#### Adding labels to the chart

```python
plt.figure(figsize=(15,5))
np.random.seed(1)
summer = np.random.uniform(25,45,size=30)
winter = np.random.uniform(10,30,size=30)
rainy = np.random.uniform(20,40,size=30)
plt.plot(summer,linestyle='-.',lw=2,color='black')
plt.plot(winter,linestyle=':',lw=2,color='red')
plt.plot(rainy,linestyle='--',lw=2,color ='green')
plt.title('Temperatures over the three seasons')
plt.xlabel('Day of the month')
plt.ylabel('Temp in degree Centigrade')
plt.legend(['summer','winter','rainy'],loc=3)
plt.show()
```

## Pie chart

```python
plt.figure(figsize=(8,8))
values = [145,400,35,130,25,250]
country = ['China','India','USA','Japan','Brazil','Pakistan']
plt.pie(values,labels=country,colors = ['blue','red','green','yellow','orange','pink'],shadow=True)
plt.title('Population Density')
plt.legend(country)
plt.show()
```

## Scatter plot

```python
plt.figure(figsize=(8,6))
weight = np.random.uniform(60,80,20)
height_2 = weight*2
#height = np.random.uniform(140,180,20)
plt.scatter(weight,height_2,marker='*',color='red')
plt.title('Weight vs Height',fontsize=20)
plt.xlabel('Weight',fontsize=15)
plt.ylabel('Height',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8,6))
weight_m = np.random.uniform(60,80,20)
weight_f = np.random.uniform(50,70,20)
height_m = weight_m*2
height_f = weight_f*1.5
plt.scatter(weight_f,height_f,marker='d',color='blue')
plt.scatter(weight_m,height_m,marker='*',color='red')
plt.title('Weight vs Height',fontsize=20)
plt.xlabel('Weight',fontsize=15)
plt.ylabel('Height',fontsize=15)
plt.legend(["Female","Male"])
plt.show()
```

## Bar chart

```python
plt.figure(figsize=(10,8))
cities = ['Delhi','Mumbai','Bangalore','Hyderabad','Chennai']
price_sqyard_in_K = [80,220,150,50,30]
plt.bar(cities,price_sqyard_in_K,color='lightblue',edgecolor='black') #bar is vertical and barh is horizontal
plt.title("Real Estate prices",fontsize=20)
plt.xlabel('Popular Cities')
plt.ylabel('Price per square yard in thousands')
plt.show()
```

### Stacked bar chart

```python
plt.figure(figsize=(10,8))
cities = ['Delhi','Mumbai','Bangalore','Hyderabad','Chennai']
price_sqyard_in_K_center = [80,220,150,50,30]
price_sqyard_in_K_outskirt = [60,100,40,20,10]
plt.bar(cities,price_sqyard_in_K_center,color='lightblue',edgecolor='black') #bar is vertical and barh is horizontal
plt.bar(cities,price_sqyard_in_K_outskirt,color='darkblue',edgecolor='black')
plt.title("Real Estate prices",fontsize=20)
plt.xlabel('Popular Cities')
plt.ylabel('Price per square yard in thousands')
plt.legend(['center','outskirt'])
plt.show()
```

### Grouped bar chart

```python
city=['Delhi','Beijing','Washington','Tokyo','Moscow']
Price_sqft_in_K_center = [60,40,70,65,15]
Price_sqft_in_K_outer = [45,22,50,52,20]
pos = np.arange(len(city))
print(pos)
w = 0.4
plt.bar(pos,Price_sqft_in_K_center,width = w,color='skyblue',edgecolor='black', label = 'City Center')
plt.bar(pos+w,Price_sqft_in_K_outer,width = w,color='darkblue',edgecolor='black', label = 'Outskirts')
plt.xticks(pos, city)
plt.xlabel('City', fontsize=16)
plt.ylabel('Price_sft_in_K', fontsize=16)
plt.title('Barchart - Price_sft_in_K across cities',fontsize=20)
plt.legend(loc = 0)
plt.show()
```

### Box and whisker plot

```python
# Salaries of different profiles in a company
fresher = np.round(np.random.uniform(4.5,6.5,size=10),1)
lead = np.round(np.random.uniform(6,10,size=10),1)
manager = np.round(np.random.uniform(8,15,size=10),1)
director = np.round(np.random.uniform(15,30,size=10),1)
cto = np.round(np.random.uniform(35,60,size=10),1)
```

```python
plt.figure(figsize=(15,10))
plt.boxplot([fresher,lead,manager,director,cto],labels=['fresher','lead','manager','director','cto'],patch_artist=True,vert=1)
plt.show()
```

```python
# sample box and whisker
data = [-10,2,3,4,5,6,6,10]
plt.boxplot(data,vert=0)
plt.axvline(2)
plt.axvline(10)
plt.axvline(2.5,color='red')
plt.axvline(4.5,color='red')
plt.axvline(6,color='red')
plt.show
```

### Histogram

```python
values = np.random.rand(100)*100
plt.figure(figsize=(10,5))
plt.hist(values,histtype='bar',edgecolor='black',bins=20)
plt.title('Histogram')
plt.show()
```

+ Line chart - time series and sequential data(numerical variable)
+ Pie chart - Compare distribution of few categories(categorical variable)
+ Bar chart - same as pie but a better version. Can compare many categories and easy to see small differences
+ Box and whisker - To check for outliers and understand the data distribution(numerical variable)
+ Scatter - To understand relation between two numerical variables i.e. checking for correlation
+ Histogram - To see distribution of numerical data(shape of data and spread of data)
